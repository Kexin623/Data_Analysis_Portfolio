```python
# The year is not correct corresponding to released so create the correct released year

def handle_releaseyear(year):
    year = str(year)
    if '(' in year:
        year = year.split(' (')[0]
        return year
    else:
        return year
    
release_year = df['released'].apply(handle_releaseyear)

def handle_releasedyear(year):
    if ', ' in year:
        year = year.split(', ')[1][:4]
        return year
    else:
        return year
released_year = release_year.apply(handle_releasedyear)
def handle_released_year(year):
    if ' ' in year:
        year = year.split()[1]
        return year
    else:
        return year
    
df['released_year'] = released_year.apply(handle_released_year)
df['released_year']
```
